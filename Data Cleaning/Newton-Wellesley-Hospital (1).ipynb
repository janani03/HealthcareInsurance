```python
import pandas as pd

file_path = '/Users/rahulrubenraj/Desktop/Newton-Wellesley-Hospital_StandardCharges.csv'

# Load the CSV file
df = pd.read_csv(file_path, encoding='latin1', low_memory=False, header=None, on_bad_lines='skip')

# Print the first 5 rows of the DataFrame
print(df.head())
```

```python
hospital_name = df.iloc[1, 0] 
license_number = df.iloc[1, 5] 
```

```python
df_data = pd.read_csv(file_path, encoding='latin1', low_memory=False, header=None, skiprows=2)

# Set the first row as the column headers
df_data.columns = df_data.iloc[0]

# Drop the first row as it is now used as the header
df_data = df_data.drop(0)

# Reset index after dropping the first row
df_data.reset_index(drop=True, inplace=True)
```

```python
df_data["hospital_name"] = hospital_name
df_data["license_number|DC"] = license_number
```

```python
columns_order = ["hospital_name", "license_number|DC"] + df_data.columns[:-2].tolist()
df_data = df_data[columns_order]
```

```python
df_data.head()
```

```python
# List of columns you want to keep
columns_to_keep = [
    'hospital_name','license_number|DC','description', 'code|2', 'code|2|type','setting' , 'standard_charge|gross', 
    'standard_charge|discounted_cash', 'payer_name', 'plan_name', 
    'standard_charge|negotiated_dollar', 'standard_charge|negotiated_percentage', 
    'standard_charge|negotiated_algorithm', 'estimated_amount', 
    'standard_charge|min', 'standard_charge|max', 'standard_charge|methodology'
]

# Keep only the desired columns
df_data = df_data[columns_to_keep]

# Print the first 5 rows to confirm
df_data
```

```python
df_data = df_data[df_data["code|2|type"] == "CPT"].reset_index(drop=True)
df_data
```

```python
df_data.info()
```

```python
# List of columns to check for 0 values
cols_to_check = [
    'standard_charge|negotiated_dollar',
    'standard_charge|negotiated_percentage',
    'standard_charge|negotiated_algorithm',
    'estimated_amount',
    'standard_charge|min',
    'standard_charge|max',
    'standard_charge|methodology'
]

# Replace empty strings and whitespace with NaN, and convert to numeric (ignoring errors)
df_data[cols_to_check] = df_data[cols_to_check].replace(r'^\s*$', '0', regex=True).apply(pd.to_numeric, errors='coerce')

# Keep rows where at least one of the selected columns is non-zero (or non-null)
df_data = df_data[~(df_data[cols_to_check].fillna(0) == 0).all(axis=1)]
```
