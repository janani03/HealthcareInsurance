```python
#importing libraries
import numpy as np
import pandas as pd
```

```python
#reading the data file
df=pd.read_csv('henry-ford-warren-hospital_standardcharges.csv',encoding='latin-1',low_memory=False)
df.head(5)
```

```python
#checking the count of null values in each col
df.isnull().sum()
```

```python
#dropping unnecessary cols
df.drop(['setting','drug_unit_of_measurement','drug_type_of_measurement','modifiers','additional_generic_notes'],axis=1, inplace= True )
```

```python
#checking the code types present in code_1
df['code|1|type'].unique()
```

```python
#checking the code types present in code_2
df['code_2_type'].unique()
```

```python
#merging CPT codes into one stanard col
df['code|2'] = df.apply(
    lambda row: row['code|1'] if row['code|1|type'] == 'CPT'
    else (row['code_2'] if row['code_2_type'] == 'CPT' else None),
    axis=1
)
```

```python
#filter CPT records alone
df_cpt = df[df['code|2'].notna()]
```

```python
#dropping code_1
df_cpt.drop(['code|1','code|1|type','code_2'],axis =1, inplace = True)
```

```python
df_cpt
```

```python
#reordering merged col
cols = list(df_cpt.columns)
# Move code|2 to index 1 (second position)
cols.insert(1, cols.pop(cols.index('code|2')))
df_cpt = df_cpt[cols]
#code_1 and code_2 CPT col were merged - the code_2_type has NaN for values coming from code_1
df_cpt['code_2_type'] = df_cpt['code_2_type'].fillna('CPT')
```

```python
df_cpt
```

```python
df_cpt = df_cpt.drop_duplicates(keep=False)
```
