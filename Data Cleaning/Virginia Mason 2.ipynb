```python
#Setting low_memory = 'False' because my columns have multiple data types
df = pd.read_csv ("Virginia Mason - Understanding Data.csv", low_memory = False, header = None)

df.head()
```

```python
new_header = df.iloc[2]  # The second row becomes the header
df = df[2:]              # Skip the first two rows (0: metadata, 1: headers)
df.columns = new_header      # Assign new header
df.reset_index(drop=True, inplace=True)

df
```
